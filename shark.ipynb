```python
lon_min, lon_max = float(meandata.longitude.min()), float(meandata.longitude.max())
lat_min, lat_max = float(meandata.latitude.min()), float(meandata.latitude.max())
```

```python
ss['Longitude'] = pd.to_numeric(ss['Longitude'], errors = 'coerce')
ss['Latitude'] = pd.to_numeric(ss['Latitude'], errors = 'coerce')
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import cartopy.crs as ccrs
```

```python
ss_within_range = ss[(ss['Longitude'] >= lon_min) & (ss['Longitude'] <= lon_max) & (ss['Latitude'] >= lat_min) & (ss['Latitude'] <= lat_max)]
ss_within_range
```

```python
plt.figure(figsize=(10,6))
plt.scatter(ss_within_range['Longitude'], ss_within_range['Latitude'])
plt.title('Scatter plot of Longitude and Latitude')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
sharklats = ss_within_range['Latitude'].values
sharklons = ss_within_range['Longitude'].values
```

```python
meandata.load()
```

```python
chlPts = []
tempPts = []
saltPts = []
for k in range(len(sharklons)):
    tmp = meandata.where(latitude=sharklats[k],longitude=sharklons[k],method='nearest')
    chlPts.append(tmp.Chl_a_sum.values+0.)
    tempPts.append(tmp.temp.values+0.)
    saltPts.append(tmp.salt.values+0.)
```

```python
ss_within_range['salt'] = saltPts
ss_within_range['temp'] = tempPts
ss_within_range['Chl_a_sum'] = chlPts
```

```python
ss_within_range.to_csv('myshark_hydrovar.csv',index=False)
```

```python
ss_within_range.rename(columns={'Latitude': 'latitude'}, inplace=True)
ss_within_range.rename(columns={'Longitude': 'longitude'}, inplace=True)
```
