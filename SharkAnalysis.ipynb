```python
year_result = int1.groupby('Incident.year').agg(
    Ave_salt=('salt', 'mean'),
    Ave_temp=('temp', 'mean'),
    Ave_Chl=('Chl_a_sum', 'mean'),
    Count=('Incident.year', 'size')
).reset_index()
```

```python
year_result
```

```python
# Create a line chart
plt.figure(figsize=(10, 6))  # Optional: Set the figure size
plt.plot(year_result['Incident.year'].values, year_result['Ave_salt'].values, label='Salt', marker='o')
plt.plot(year_result['Incident.year'].values, year_result['Ave_Chl'].values, label='Chl', marker='s')
plt.plot(year_result['Incident.year'].values, year_result['Ave_temp'].values, label='Temp', marker='^')
plt.xlabel('Years')
plt.ylabel('Values')
plt.title('Line Chart for Salt, Chl, and Temp')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calculate the correlation matrix
last_four_columns = year_result.iloc[:, -4:]
correlation_matrix = last_four_columns.corr()
plt.figure(figsize=(8, 6))  # Optional: Set the figure size
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.xlabel('Features')
plt.ylabel('Features')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
low_3 = year_result.sort_values(by='Count').head(3)
top_3 = year_result.sort_values(by='Count', ascending=False).head(3)
low_3_mean = low_3[['Ave_salt', 'Ave_temp', 'Ave_Chl']].mean()
top_3_mean = top_3[['Ave_salt', 'Ave_temp', 'Ave_Chl']].mean()
mean_df = pd.DataFrame({'Low_3_Mean': low_3_mean, 'Top_3_Mean': top_3_mean})
```

```python
# Create a bar chart for the mean DataFrame
ax = mean_df.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Three variables')
plt.ylabel('Mean Value')
plt.title('Difference between high attack and low attack')
plt.legend(['High Attack', 'Low Attack'])
ax.set_xticklabels(ax.get_xticklabels(), rotation=0)
# Show the bar chart
plt.show()
```
